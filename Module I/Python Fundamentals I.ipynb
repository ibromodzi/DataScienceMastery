### Expressions, Operators and Precedence in Python

#### Expressions
An expression is made with combinations of variables, values, operators and function calls. The Python interpreter evaluates the valid expression.

#### Arithmetic Operators
1. Multiplication *
2. Addition	      +
3. Subtraction    -
4. Division	      /
5. Modulus        %
6. Exponent       **
7. Assigments     =
8. We also have other types of operators such as logical operators, comparison operators, bitwise operators etc..

#### Precedence
Programming languages follow a strict rule on how compound expressions, such as 4 + 5 * 3, are evaluated. The order in which expressions are evaluated is as follow;
* Parenthesis --> Exponentiation --> Multiplication --> Division --> Addition --> Subtraction

```python
# A simple expression
20 + 3
```

```python
4 + 5 * 3
```

5 multiply by 3 is evaluated first because multiplication has a higher precedence, the result is then added to 4

```python
(4 + 5) * 3
```

By using parenthesis, we are forcing the interpreter to first evaluate the expression inside

```python
((4 + 5) * 3) / 3
```

```python
4 + 5 * 3 /3
```

# Making your program interactive
We can make our program interactive by getting information from the user(s) of our program. By using the `input()` function, the user is prompted to enter an information. The program will essentially prompt the user for information, and then when the user hits enter, will take whatever they entered and be able to use it going forward in it's execution.

It is important to note that whatever the user enters is stored in a variable and then we can further use it in our program.

```python
# get user name
name = input("Please enter your name: ")
```

```python
print("You are welcome to our program", name +"!")
```

```python
# get the age of the user
age = int(input("What's your age? "))
```

```python
futureAge = age + 5
```

```python
print("You'll be", futureAge, "years older in five years")
```

```python
print("You'll be", age + 5, "years older in five years")
```
