## Introduction to Data Visualization with Matplotlib

In the preceding sections, you've gained a solid foundation in Python programming, explored object-oriented programming (OOP), manipulated data using NumPy, and conducted data analysis with Pandas. Now, we're diving into the exciting world of data visualization with Matplotlib.

**Matplotlib** is a versatile and powerful library in Python for creating a wide range of static and interactive visualizations. Effective data visualization is essential in data analysis, helping us uncover patterns, trends, and insights in our data. Whether you're exploring data during the initial stages of analysis or presenting findings in a report or presentation, Matplotlib is a valuable tool in your data science toolkit.

### Why Matplotlib?

You might ask, "Why do we need Matplotlib when there are other visualization libraries?" Matplotlib is chosen for its flexibility and ubiquity. It provides fine-grained control over every aspect of your plots, making it suitable for a wide range of visualization needs. In this module, we'll explore various chart types that Matplotlib offers, each serving specific purposes:

- **Line Charts:** Visualize trends and patterns over time or a continuous range.
- **Histograms:** Understand data distributions and spot outliers.
- **Bar Charts:** Compare categorical data or show relationships between variables.
- **Pie Charts:** Display parts of a whole.
- **Scatter Plots:** Explore relationships and correlations between variables.
- **Box Plots:** Visualize summary statistics and detect outliers.

### Key Concepts in Matplotlib

As we journey through Matplotlib, we'll cover essential concepts and functionalities, including:

- **Creating Basic Plots:** Learn the fundamentals of creating simple plots.
- **Customizing Aesthetics:** Fine-tune plot elements such as colors, labels, and legends.
- **Multiple Subplots:** Create complex layouts with multiple subplots.
- **Saving and Exporting:** Save your visualizations in various formats.

Most of the time we have to work with the Pyplot as an interface of Matplotlib. So, we import Pyplot like this:

```python
import matplotlib.pyplot as plt
import pandas as pd

# prepare the dataset
hap_df = pd.read_csv("world_happines.csv")
hap_df.info()
```

```python
hap_df.head(5)
```

```python
# Let's make a simple line plot with this data
# we want to see the relationship between a country's happines score and tir life expectancy

# we first need to create a series of the two variables
score = hap_df.Score
expec = hap_df["Healthy_Life_Expectancy"]

plt.plot(score, expec)
plt.show()
```

**The first argument, score, will be plotted on the x-axis, and the second, expec, will be plotted on the y-axis. Calling plt.show() is required for your graph to actually be printed onto your screen**

```python
# customizing the charts more
plt.plot(score, expec)
plt.title('Happiness score vs Life expectance')
plt.xlabel('Happiness Score')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
# We want to know the relationship betwen rank of the country and their GDP & Life expectancy
gdp = hap_df['GDP_Per_Capita']
rank = hap_df.Rank

plt.plot(rank, gdp, label="gdp")
plt.plot(rank, expec, label="life expct")
plt.title('Rank Vs GDP & Life expectnacy')
plt.xlabel('Rank')
plt.legend(loc="upper right")
plt.show()
```

#### Bar chart
A bar chart or bar graph is a chart or graph that presents categorical data with rectangular
bars with heights or lengths proportional to the values that they represent. The bars can
be plotted vertically or horizontally. 

```python
# Let's first get the first eight countries with highest GDP and plot the bar graph
country_by_gdp = hap_df.sort_values(by="GDP_Per_Capita",ascending=False).head(8)
country_by_gdp
```

```python
country = country_by_gdp["Country"]
gdp = country_by_gdp["GDP_Per_Capita"]
plt.bar(country, gdp)
plt.title("Country by GDP")
plt.xlabel("Country")
plt.ylabel("GDP")
plt.show()
```

```python
# We can make the xticks better by rotating it 90 degrees
country = country_by_gdp["Country"]
gdp = country_by_gdp["GDP_Per_Capita"]
plt.bar(country, gdp)
# we'll specify more font properties here
plt.title("Country by GDP", fontsize=12, fontfamily="Times New Roman")
plt.xlabel("Country")
plt.ylabel("GDP")
plt.xticks(rotation=90)
plt.show()
```

```python
# Again, let's sor by gdp, life expectancy and happiness score to create a multiple bar chart
country_by_metrics = hap_df.sort_values([ "Score", "GDP_Per_Capita", "Healthy_Life_Expectancy"], 
                                        ascending = False).head(8)
country_by_metrics
```

**Note: It is important to understand that when you are sorting by multiple columns, the sort sequence is exactly as specified. In other words, using the output in the example above, the country are sorted by the gdp column only when multiple rows have the same score value. If all the values in the score column had been unique, no data would have been sorted by gdp.**

```python
import numpy as np
X = np.arange(len(country_by_metrics))
xticks_label = country_by_metrics["Country"]

plt.bar(X + 0.00, country_by_metrics["Score"], width=0.25, color='r')
plt.bar(X + 0.25, country_by_metrics["GDP_Per_Capita"], width=0.25, color="b")
plt.bar(X + 0.50, country_by_metrics["Healthy_Life_Expectancy"], width=0.25, color='g')
plt.title("Country vy GDP, Happiness score, and life expectancy")
plt.xlabel("Country")
plt.xticks(X, xticks_label, rotation=90)
plt.legend(labels=["score", "gdp", "expectancy"])
plt.show()
```

#### Histogram
A histogram is an accurate representation of the distribution of numerical data. It is an
estimate of the probability distribution of a continuous variable.<br>
To construct a histogram, follow these steps:
* Bin the range of values.
* Divide the entire range of values into a series of intervals.
* Count how many values fall into each interval. 

```python
titanic_df = pd.read_csv("titanic.csv")
```

```python
# Let's plot the age distribution of passengers from our titanic df
titanic_df.head(4)
```

```python
# let's filter cut out the Age row
age = titanic_df.Age
bins =  [0,25,50,75,100]
plt.hist(age, bins)
plt.title("Age distribution of passengers")
plt.xlabel("Age")
plt.ylabel("Number of passenger(n)")
plt.show()
```

#### Pie Chart
A Pie Chart can only display one series of data. Pie charts show the size of items (called
wedge) in one data series, proportional to the sum of the items

```python
pclass = titanic_df["Pclass"].value_counts().reset_index()
pclass
```

```python
pass_class = pclass["index"]
frequecy = pclass["Pclass"]
plt.pie(frequecy, labels=pass_class, autopct='%1.2f%%')
plt.title("Passenger class distribution")
plt.show()
```

#### Scatter Plot
Scatter plots are used to plot data points on horizontal and vertical axis in the attempt to
show how much one variable is affected by another.

```python
# Let's see the relationship between GDP and Happiness of the countries
gdp = hap_df["GDP_Per_Capita"]
score = hap_df["Score"]
plt.scatter(gdp, score)
plt.title("Relationship between Happiness and GDP of Countries")
plt.xlabel("GDP")
plt.ylabel("Happiness score")
plt.show()

# we can as well save our charts using plt.savefig() method
plt.savefig("scatter.png")
```

#### Box Plot
A box plot which is also known as a whisker plot displays a summary of a set of data
containing the minimum, first quartile, median, third quartile, and maximum. In a box
plot, we draw a box from the first quartile to the third quartile. A vertical line goes through
the box at the median. The whiskers go from each quartile to the minimum or maximum<br>

Plotting box plots is more straight forward using pandas

```python
titanic_df["Sex"].value_counts().plot(kind="box")
```

### Hints for Selecting the Right Chart

Choosing the right type of chart is a critical step in effective data visualization. The choice depends on the nature of your data and the message you want to convey. Here are some hints to help you select the most suitable chart:

#### Line Charts:
- **Use**: Show trends, changes over time, or continuous data.
- **Example**: Time-series data, stock prices, temperature fluctuations.

#### Histograms:
- **Use**: Visualize the distribution and frequency of numerical data.
- **Example**: Age distribution, exam scores, income distribution.

#### Bar Charts:
- **Use**: Compare categories or display data across different groups.
- **Example**: Sales by product category, population by city.

#### Pie Charts:
- **Use**: Show parts of a whole and the contribution of each category.
- **Example**: Market share by company, composition of a budget.

#### Scatter Plots:
- **Use**: Explore relationships and correlations between two numerical variables.
- **Example**: Scatter plot of height vs. weight, GDP vs. life expectancy.

#### Box Plots:
- **Use**: Visualize the distribution and summary statistics of numerical data, especially for comparing multiple groups.
- **Example**: Compare the ages of passengers by class in the Titanic dataset.

Remember that the effectiveness of a chart also depends on its design and labeling. Ensure that your visualizations are clear, labeled, and convey your intended message.
